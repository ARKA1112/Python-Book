#### Classes Generate Multiple Instance Objects

```python
#two kinds of objects in python oop model
#class objects and instance objects

#classes provide default behaviour and serv4e as factories for instance objects.

#calling the class objects create instance objects
#the instance objects inherit all the attributes of the class

#the first argument usually self references the instance object being processed assignments to attributes of self create or change dat in the instance not the class.

```

```python
#First class example

class FirstClass:                   #Define a class object
    def setdata(self,value):        #Define class method
        self.data = value           #self  is the instance
    def display(self):              
        print(self.data)            #self.data per instance
```

```python
x = FirstClass()    #two instances created
y = FirstClass()    #Each is a new namespace
#each have access to the namespaces or attributes of the class named FirstClass()
```

```python
#x.display() will not work until setname has been assigned
#hence 

x.setdata('New Value')
x.display()
```

#### A Second Example


```python
class secondclass(FirstClass):            #inherits firstclass attributes
    def display(self):
        print('Current value = "%s"'%self.data)
```

```python
z = secondclass() #instance of second class
z.setdata("inherits from the first class") #setdata inherited from the firstclass
z.display()  #displays the value
```

#### Operator overloading


```python
#__init__ object is run when a new instance object is created
#ENABLES TO RUN THE CLASS AS A FUNCTION/ TO TAKE INPUT/SET THE ARGUMENT AT RUNTIME ISTEAD OF FETCHING THE INPUT CLASS FROM THE SUPERCLASS

#__add__ is run when a ThirdClass instance appears in a + expression ENABLES TO ADD AN OPERATION TO ADD SOMETHING WITH THE RETURN

#__str__ is run when an object is printed(technically it converts the object into a string) MAINLY USED FOR RETURNING THE OUTPUT AS PRINT 

#For example

class ThirdClass(secondclass):
    def __init__(self,value):
        self.data = value
    def __add__(self,other):
        return ThirdClass(self.data + other)
    def __str__(self):
        return '[ThirdClass: %s]' % self.data
    def mul(self, other):
        self.data *= other
```

```python
a = ThirdClass('abc')
a.display()
print(a)
b = a + 'xyz' #creates a new instance because __add__ return the ThirdClass when called hence b also becomes the instance of ThirdClass

#It can also be done without returning the ThirdClass insted we can return self.data + other. Now the return will be a string and b will not be instance instead it will be a string
b.display()
```

```python
b.mul(4)
b.display()
print(b)
```
