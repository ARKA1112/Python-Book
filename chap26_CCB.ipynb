#### Classes Generate Multiple Instance Objects

```python
#two kinds of objects in python oop model
#class objects and instance objects

#classes provide default behaviour and serv4e as factories for instance objects.

#calling the class objects create instance objects
#the instance objects inherit all the attributes of the class

#the first argument usually self references the instance object being processed assignments to attributes of self create or change dat in the instance not the class.

```

```python
#First class example

class FirstClass:                   #Define a class object
    def setdata(self,value):        #Define class method
        self.data = value           #self  is the instance
    def display(self):              
        print(self.data)            #self.data per instance
```

```python
x = FirstClass()    #two instances created
y = FirstClass()    #Each is a new namespace
#each have access to the namespaces or attributes of the class named FirstClass()
```

```python
#x.display() will not work until setname has been assigned
#hence 

x.setdata('New Value')
x.display()
```

```python
x.setdata('arka')
x.display()
```

#### Classes are customized by inheritance

```python
#to inherit attributes of other classes simply pass them into parentheses

#classes inherit their attributes from their superclasses

#instances inherit attr from all accessible classes
```

#### A Second Example


```python
class secondclass(FirstClass):            #inherits firstclass attributes
    def display(self):
        print('Current value = "%s"'%self.data)
```

```python
z = secondclass() #instance of second class
z.setdata("inherits from the first class") #setdata inherited from the firstclass
z.display()  #displays the value
```
